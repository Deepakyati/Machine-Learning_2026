```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset('iris')
```

```python
df.head()
```

```python
df['species'].unique()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.drop('species',axis=1)
y=df['species']
```

```python
x
```

```python
x_train,x_test, y_train,y_test=train_test_split(x,y,test_size=0.33, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model_knn=KNeighborsClassifier(n_neighbors=5)
```

```python
model_knn.fit(x_train,y_train)
```

```python
model_knn.score(x_test,y_test)
```

```python
#n=9

model_knn=KNeighborsClassifier(n_neighbors=9)
```

```python
model_knn.fit(x_train,y_train)
```

```python
model_knn.score(x_test,y_test)
```

```python
#n=13

model_knn=KNeighborsClassifier(n_neighbors=13)
```

```python
model_knn.fit(x_train,y_train)
```

```python
model_knn.score(x_test,y_test)
```

# 100% accuracy means overfit

# SVM

```python
from sklearn.svm import SVC
```

```python
model_svm=SVC(C=30,kernel='rbf', gamma='auto')
```

```python
model_svm.fit(x_train,y_train)
```

```python
model_svm.score(x_test,y_test)
```

```python
#kernel=linear

model_svm=SVC(C=30,kernel='linear', gamma='auto')

model_svm.fit(x_train,y_train)

model_svm.score(x_test,y_test)
```

```python
#C=10

model_svm=SVC(C=10,kernel='linear', gamma='auto')

model_svm.fit(x_train,y_train)

model_svm.score(x_test,y_test)
```

# Grid Search CV

```python
from sklearn.model_selection import GridSearchCV
```

```python
classifier = GridSearchCV((model_svm),{
    'C':[1,10,20,30],
    'kernel':['rbf','linear']

},cv=5, return_train_score=False)
```

```python
classifier.fit(x,y)
```

```python
classifier.cv_results_
```

```python
results=pd.DataFrame(classifier.cv_results_)
```

```python
results
```

```python
results[['param_C',	'param_kernel', 'mean_test_score']]
```

# Random Search CV

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
classifier_r = RandomizedSearchCV((model_svm),{
    'C':[1,10,20,30],
    'kernel':['rbf','linear']

},n_iter=4,cv=5, return_train_score=False)
```

```python
classifier_r.fit(x,y)
```

```python
result = pd.DataFrame(classifier_r.cv_results_)
```

```python
result
```

```python
result[['param_C',	'param_kernel', 'mean_test_score']]
```
